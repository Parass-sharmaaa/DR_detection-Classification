```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def predict_class(path):
    # Attempt to read the image
    img = cv2.imread(path)

    # Check if the image was loaded successfully
    if img is None:
        print(f"Error: Unable to load image at path: {path}")
        return

    # Convert the image to RGB and resize
    RGBImg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    RGBImg = cv2.resize(RGBImg, (224, 224))
    plt.imshow(RGBImg)

    # Preprocess the image for model prediction
    image = np.array(RGBImg) / 255.0
    image = np.expand_dims(image, axis=0)

    # Load the pre-trained model
    model = tf.keras.models.load_model('/content/drive/MyDrive/Trained Model/DRmodel.keras')

    # Predict and interpret the output
    predict = model.predict(image)
    predicted_class = np.argmax(predict, axis=1)[0]

    if predicted_class == 0:
        print('Prediction: No_DR')
    else:
        print('Prediction: DR')
```
